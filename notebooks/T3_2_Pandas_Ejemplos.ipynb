**Transformación y agregación de datos en Pandas**

Para ilustrar el proceso de limpieza y preparación de datos, vamos a usar un nuevo dataset, que describe características y precios promedio de viviendas en cada distrito de California.

```python
url = "https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/housing.csv"
housing_df = pd.read_csv(url)
housing_df.head()
```

```python
housing_df.info()
housing_df.describe()
```

```python
# Obtiene recuentos de los posibles valores en un atributo categórico
housing_df["ocean_proximity"].value_counts()
```

```python
# Manejo de valores perdidos
housing_df.isna().sum()  # Detección
```

```python
# Opción 1: eliminar filas con valores faltantes
housing_df_dropna = housing_df.dropna(subset=["total_bedrooms"])
housing_df_dropna
housing_df_dropna.shape
```

```python
# Opción 2: imputar con la mediana
mediana = housing_df["total_bedrooms"].median()
housing_df["total_bedrooms"] = housing_df["total_bedrooms"].fillna(mediana)
housing_df
housing_df.shape
```

```python
# Eliminación de duplicados
housing_df.drop_duplicates(inplace=True)
housing_df.shape
```

**One-hot encoding**, mediante la función `pd.get_dummies()` para la conversión de un atributo categórico en varios atributos numéricos binarios, uno para cada posible categoría en el atributo original.

```python
df_dummies = pd.get_dummies(housing_df, columns=["ocean_proximity"])
df_dummies
```

**Ingeniería de características**: agregar columnas y transformar nuevas columnas a partir de otras

```python
# Relación/ratio de habitaciones por hogar
housing_df["rooms_per_household"] = housing_df["total_rooms"] / housing_df["households"]

# Relación/ratio población media por hogar
housing_df["population_per_household"] = housing_df["population"] / housing_df["households"]
housing_df.head()
```

```python
# Renombrar una columna
housing_df.rename(columns={"housing_median_age": "median_age"}, inplace=True)

# Eliminar columnas
housing_df.drop(columns=["longitude", "latitude"], inplace=True)
housing_df.head()
```

```python
# Promedio de ingresos por categoría del atributo de proximidad al océano
housing_df.groupby("ocean_proximity").mean()[["median_income", "median_house_value"]]
```

**ESCALADO DE ATRIBUTOS**

```python
# Guardar en un D.F. separado los atributos numéricos
atributos_numericos = housing_df.select_dtypes(include=['float64', 'int64']).columns
housing_df_num = housing_df[atributos_numericos]
```

```python
# Normalización manual (Min-Max) con Pandas.
# Para cada valor x de un atributo, se hace: (x - min) / (max-min)

df_manual_minmax = (housing_df_num - housing_df_num.min()) / (housing_df_num.max() - housing_df_num.min())
df_manual_minmax
```

```python
# Estandarización (z-score) manual.

df_standardized = (housing_df_num - housing_df_num.mean()) / housing_df_num.std()
df_standardized
```

Aunque Pandas permite el escalado manual, a menudo es preferible en la práctica realizar el escalado con otras librerías como **scikit-learn*, ya que dispone de funcionar bien probadas y eficientes.

```python
# Estandarización (z-score) con scikit-learn
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(housing_df_num), columns=atributos_numericos)
# Cuidado: incluso especificando solo el escalado a ciertos atributos numéricos
# el hecho de pasar un df que pueda contener categóricos como 'housing_df' va a dar error
df_standardized
```

```python
# Normalización Min-Max con scikit-learn

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_minmax = pd.DataFrame(scaler.fit_transform(housing_df_num), columns=atributos_numericos)
df_minmax
```

```python
# Escalado robusto, usando mediana y rango intercuartílico
# Es menos robusto a outliers

from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df_robust = pd.DataFrame(scaler.fit_transform(housing_df_num), columns=atributos_numericos)
df_robust
```

```python
import numpy as np
# Transformación logarítmica:
# Especialmente útil cuando los datos están fuertemente sesgados a la derecha,
# esto es, una distribución asimétrica con una cola larga hacia valores grandes.

df_log = np.log1p(housing_df_num)
df_log
```

Aunque aquí hemos probado varios posibles enfoques de escalado sobre todos los atributos numéricos del dataset, aplicar estas transformaciones "a diestro y siniestro" sobre todos los atributos numéricos puede no solo ser innecesario, sino a vces contraproducente dependiendo de la naturaleza de los datos y problema de análisis.

```python
# Paso 1: Eliminar las columnas numéricas originales del DataFrame completo
df_no_numeric = housing_df.drop(columns=housing_df_num.columns)

# Paso 2: Concatenar el DataFrame de columnas numéricas escaladas con el DataFrame de atributos categóricos
df_final = pd.concat([df_no_numeric, housing_df_num], axis=1)
df_final
```

**Detección y manejo de outliers**

```python
import pandas as pd

# Cargar el dataset
url = 'https://raw.githubusercontent.com/gakudo-ai/open-datasets/main/50_Startups.csv'
df = pd.read_csv(url)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot para la columna 'Profit' (se puede hacer horizontal o vertical con argumento x/y)
sns.boxplot(y=df['Profit'])
plt.show()
```

```python
# Calcular Q1 (25%) y Q3 (75%)
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)

# Calcular el IQR
IQR = Q3 - Q1

# Definir los límites inferior y superior
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filtrar los datos sin outliers
df_filtered = df[(df['Profit'] >= lower_limit) & (df['Profit'] <= upper_limit)]
df_filtered.shape
```

Método alternativo, a partir de un *scatterplot* para la detección

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un scatterplot entre 'R&D Spend' y 'Profit'
sns.scatterplot(data=df, x='R&D Spend', y='Profit')
plt.title('Scatterplot de R&D Spend vs Profit')
plt.show()
```

```python
# Definir un umbral de valores para 'R&D Spend' y 'Profit' (basado en observación visual del gráfico)
# Aquí, consideramos un rango razonable para ambas variables

df_filtered = df[(df['R&D Spend'] > 5000) & (df['Profit'] > 25000)]

# Visualizar el nuevo scatterplot después de eliminar los outliers
sns.scatterplot(data=df_filtered, x='R&D Spend', y='Profit')
plt.title('Scatterplot de R&D Spend vs Profit (sin outliers)')
plt.show()
```
