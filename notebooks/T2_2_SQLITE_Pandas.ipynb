```python
import pandas as pd
import sqlite3

# 1. Conexión a nuestro archivo de BBDD existente
conn = sqlite3.connect('cartera_financiera.db')

# 2. Consulta a la tabla 'Clientes' buscando perfiles de riesgo (score bajo)
query = "SELECT Empresa, Sector, Score_Crediticio FROM Clientes WHERE Score_Crediticio < 0.7"

# 3. Volcado de los datos a un DataFrame, ¡listo para un proceso de ML!
df_riesgo = pd.read_sql_query(query, conn)
df_riesgo.head()
```
