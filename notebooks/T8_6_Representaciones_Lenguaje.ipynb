Este taller contiene ejemplos para familiarizarse con diferentes formas de representar el lenguaje natural con librerías Python.

```python
texts = [
"I loved this movie, it was fantastic!",
"The movie was boring and too long.",
"Absolutely wonderful acting and great story.",
"Terrible film, I will not watch it again.",
]
```

```python
# Token IDs con Keras Tokenizer

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>") # Token especial para representar palabras fuera de vocabulario, y no perderlas
tokenizer.fit_on_texts(texts) # Analiza el corpus de texto y construye el vocabulario. Prepara el tokenizador


word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(texts) # Convertir cada texto en una secuencia de token IDs
padded = pad_sequences(sequences, maxlen=10, padding='post') # Se añaden los valores de Padding al final, a la derecha del vector de token IDs


print("Word Index:", word_index)
print("Sequences:", sequences)
print("Padded Sequences:\n", padded)
```

```python
# Representación Bag of Words

from sklearn.feature_extraction.text import CountVectorizer


bow = CountVectorizer()
bow_matrix = bow.fit_transform(texts)


print("Vocabulary:", bow.get_feature_names_out())
print("BoW Matrix:\n", bow_matrix.toarray())
```

```python
# Representación TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer


tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(texts)


print("TF-IDF Vocabulary:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
# Word embeddings (Word2Vec)

from gensim.models import Word2Vec
from gensim.utils import simple_preprocess


processed = [simple_preprocess(t) for t in texts]
w2v_model = Word2Vec(sentences=processed, vector_size=50, window=5, min_count=1)
# sentences es una lista de listas de tokens (una lista de tokens por secuencia)
# Dimension del vector embedding usado para codificar cada token
# Window: tamaño de contexto, num. palabras a mirar a ambos lados, ya que Word2Vec aprende relaciones de co-ocurrencia
# Ignora palabras que aparecen menos de min_count veces, evitando hacer embeddings de palabras extremadamente raras

print("Vector for 'movie':", w2v_model.wv['movie'])
```

Los llamados **Sentence Transformers** son capaces de construir representaciones vectoriales de frases completas, es decir, embeddings de frases, párrafos o documentos completos. Frases similares tendrán embeddings cercanos entre sí.

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
sentence_embeddings = model.encode(texts)
print("Sentence Embeddings shape:", sentence_embeddings.shape)

sentence_embeddings
```
