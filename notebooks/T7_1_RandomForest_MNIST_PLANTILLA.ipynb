Este ejemplo utiliza *scikit-learn* para construir un ensemble de bosque aleatorio, **Bosque Aleatorio** (múltiples modelos de árboles de decisión construidos mediante muestreo de atributos).

El dataset que se utilizará es el popular conjunto de datos **MNIST** para el reconocimiento de dígitos manuscritos. Las instancias son imágenes de dígitos de 8x8 píxeles. El conjunto de datos tiene **64 atributos predictores** (uno por cada píxel de la imagen) y un atributo de clase con **10 clases posibles** asociadas a los dígitos del 0 al 9, es decir, el píxel visualizado en la imagen de 8x8 píxeles.
