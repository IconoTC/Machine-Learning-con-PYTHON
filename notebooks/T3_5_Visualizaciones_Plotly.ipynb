```python
import pandas as pd
import plotly.express as px # API de alto nivel de Plotly

# Cargar datos
penguins = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/penguins.csv')

# Crear gráfico de dispersión
fig = px.scatter(penguins,
                 x='bill_length_mm',
                 y='bill_depth_mm',
                 color='species',
                 # También se pueden colorear los puntos según un atributo numérico, p.ej. 'flipper_length_mm'
                 # Opcional:
                 # hover_data=['flipper_length_mm', 'body_mass_g'],
                 # hover_name='island',
                 title='Relación entre longitud y profundidad del pico por especie')
fig.show()
```

```python
# Crear gráfico de barras
fig = px.histogram(penguins,
                   x='island',
                   labels={'island':'Nombre de la Isla'},
                   # OPCIONAL:
                   # color='species',
                   # barmode='group', # las opciones son 'stack' y 'group'
                   title='Distribución de pingüinos por isla')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

cta = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/CTA_-_Ridership_-_Daily_Boarding_Totals.csv')
# Alternativa: https://data.cityofchicago.org/api/views/6iiy-9s97/rows.csv?accessType=DOWNLOAD

# La columna de fecha debe estar en formato datetime
cta['service_date'] = pd.to_datetime(cta['service_date'])

start_date = pd.to_datetime('2018-01-01')
cta_recent = cta[cta['service_date'] >= start_date].copy()
cta_recent = cta_recent.sort_values(['service_date'])

# Crear gráfico de líneas del número de embarques en transporte público
fig = px.line(cta_recent,
              x='service_date',
              y='total_rides',
              title='Uso diario del transporte público en Chicago')
fig.show()
```

```python
# este dataset parece tener algunos registros inconsistentes o desordenados.
# El siguiente agrupamiento ayuda a solventarlo.
daily_total_rides = cta_recent.groupby('service_date')[['total_rides','bus','rail_boardings']].sum().reset_index()

# Crear gráfico de líneas del número de embarques totales diarios
fig = px.line(daily_total_rides,
              x='service_date',
              y='total_rides',
              title='Uso diario del transporte público en Chicago')
fig.show()
```

```python
# AÑADIDO DE OPCIONES DE PERSONALIZACIÓN EXTRA AL GRÁFICO ANTERIOR
fig = px.line(cta_recent,
              x='service_date',
              y='total_rides',
              hover_data=['bus','rail_boardings'],
              title='Uso diario del transporte público en Chicago')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Cargar datos
titanic = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/titanic.csv')

# Boxplot de edad separado para cada clase social
fig = px.box(titanic,
             x='Pclass',
             y='Age',
             title='Distribución de edades por clase en el Titanic')
fig.show()
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/titanic.csv')

# Boxplot de edad separado para cada clase social
fig = px.box(titanic,
             x='Pclass',
             y='Age',
             color='Sex',
             # orientation='h', # Si se usa, hay que intercambiar x e y
             labels={'Pclass':'Clase Social', 'Age':'Edad'},
             hover_data=['SibSp', 'Parch'],
             title='Distribución de edades por clase en el Titanic')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Cargar datos
startups = pd.read_csv('https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/50_Startups.csv')

# Crear gráfico de dispersión
fig = px.scatter(startups,
                 x='R&D Spend',
                 y='Profit',
                 color='State',              # Coloreamos los puntos según el valor de 'State'
                 size='Marketing Spend',     # Ajustamos el tamaño de los puntos según 'Marketing Spend'
                 hover_name='State',          # Mostrar el estado como título al pasar el cursor
                 hover_data=['Administration'], # Aañadir datos de 'Administration' al pasar el cursor
                 log_x=False,                 # La escala logarítmica puede ser útil cuando hay outliers
                 size_max=20,                # Máximo tamaño de los puntos
                 title='Relación entre gasto en I+D y ganancias por Estado, con tamaño según Gasto de Marketing'
                )

fig.update_layout(width=800)
fig.show()
```
